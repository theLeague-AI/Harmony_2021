## The Data

```python
# Replacing missing values

mobile_data_raw['n_cores'].value_counts()
mobile_data_raw['int_memory'].value_counts()

clockspeed_sim = dict(round(mobile_data_raw.groupby('n_cores')['clock_speed'].mean()))
ram_sim = dict(round(mobile_data_raw.groupby('int_memory')['ram'].mean()))
```

```python
mobile_data_raw['clockspeed_mean'] = mobile_data_raw['n_cores'].apply(lambda x: clockspeed_sim[x])
mobile_data_raw['clock_speed'].fillna(mobile_data_raw['clockspeed_mean'], inplace=True)

mobile_data_raw['ram_mean'] = mobile_data_raw['int_memory'].apply(lambda x: ram_sim[x])
mobile_data_raw['ram'].fillna(mobile_data_raw['ram_mean'], inplace=True)


mobile_data_raw.isnull().sum()
```

```python
# Applying feature scaling

#ram -pc -int_memory -fc -clock_speed

# Find the max and min from the ram column
max_ram = mobile_data_raw.ram.max()
min_ram = mobile_data_raw.ram.min()

max_pc = mobile_data_raw.pc.max()
min_pc = mobile_data_raw.pc.min()


max_fc = mobile_data_raw.fc.max()
min_fc = mobile_data_raw.fc.min()

max_int_memory = mobile_data_raw.int_memory.max()
min_int_memory = mobile_data_raw.int_memory.min()

max_clock_speed = mobile_data_raw.clock_speed.max()
min_clock_speed = mobile_data_raw.clock_speed.min()

mobile_data_raw['ram'] = mobile_data_raw.ram.map(lambda p: (p - min_ram)/(max_ram - min_ram))
mobile_data_raw['pc'] = mobile_data_raw.pc.map(lambda p: (p - min_pc)/(max_pc - min_pc))
mobile_data_raw['int_memory'] = mobile_data_raw.pc.map(lambda p: (p - min_int_memory)/(max_int_memory - min_int_memory))
mobile_data_raw['clock_speed'] = mobile_data_raw.pc.map(lambda p: (p - min_clock_speed)/(max_clock_speed - min_clock_speed))

mobile_data_raw.head()
```

```python
import seaborn as sns

def correlation_heatmap(mobile_data_raw):
    correlations = mobile_data_raw.corr()

    fig, ax = plt.subplots(figsize=(20,20))
    sns.heatmap(correlations, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .70})
    plt.show();
    
correlation_heatmap(mobile_data_raw)
```

## Building the Model

Now that the data has been processed it is time to determine and build the model that will be used to find our predictions. 

---

Consider the following points before making a choosing a model:

- Create a train and test sets of data from the provided data

```python
# Code to build Model
mobile_data_raw = mobile_data_raw.drop(columns=["px_width"])
target = mobile_data_raw["price_range"]
features = mobile_data_raw.drop('price_range', axis = 1)
```

```python
features.info()
```

```python
mobile_data_raw.info()
```

```python
mobile_data_raw.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
model = RandomForestClassifier()

model.fit(X_train, y_train)

feature_weights = model.feature_importances_
feature_weights

indices = np.argsort(feature_weights)[::-1]
columns = X_train.columns.values[indices[:5]]
values = feature_weights[indices][:5]

fig = plt.figure(figsize=(9, 5))
plt.title("Normalized Weights for First Five Most Predictive Features", fontsize=16)
plt.bar(np.arange(5), values, width=0.6, align="center", color='#93389d', \
       label="Feature Weight")
plt.bar(np.arange(5) - 0.3, np.cumsum(values), width=0.2, align="center", color='#ff4500', \
       label="Cumulative Feature Weight")
plt.xticks(np.arange(5), columns)
plt.xlim((-0.5, 4.5))
plt.ylabel("Weight", fontsize=12)
plt.xlabel("Feature", fontsize=12)

plt.legend(loc='upper center')
plt.tight_layout()
plt.show()  
```

```python
# Code to find Accuracy metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, balanced_accuracy_score 
from sklearn.metrics import precision_score, recall_score
mobiledata_confusion_matrix = confusion_matrix(y_test, y_test_pred)
print(mobiledata_confusion_matrix)
```

```python
from sklearn.metrics import balanced_accuracy_score, precision_score, recall_score

train_accuracy = balanced_accuracy_score(y_train, y_train_pred)
test_accuracy = balanced_accuracy_score(y_test, y_test_pred)
train_precision = precision_score(y_train, y_train_pred)
test_precision = precision_score(y_test, y_test_pred)
train_recall = recall_score(y_train, y_train_pred)
test_recall = recall_score(y_test, y_test_pred)

print('The training accuracy is {}%'.format(train_accuracy*100))
print('The training precision is {}%'.format(train_precision*100))
print('The training recall is {}%'.format(train_recall*100))
print('The test accuracy is {}%'.format(test_accuracy * 100))
print('The test precision is {}%'.format(test_precision * 100))
print('The test recall is {}%'.format(test_recall * 100))
```
